# Bar chart for furnishing status distribution

```python
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\HP\Downloads\archive (15)\Housing.csv')
furnishing_counts = df['furnishingstatus'].value_counts()
plt.figure(figsize=(8, 6))
furnishing_counts.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Distribution of Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Number of Houses')
plt.xticks(rotation=0)
plt.show()
```
